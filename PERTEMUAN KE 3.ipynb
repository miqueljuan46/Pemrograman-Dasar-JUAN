```python
# Program Kasir Sederhana Kedai Kopi

def kasir_kedai_kopi(harga_item, is_member):
    subtotal = sum(harga_item)
    # Evaluasi Syarat Diskon:
    if is_member and subtotal > 100000:
        diskon = 0.15 * subtotal
    else:
        diskon = 0
    # Hitung Harga Setelah Diskon
    harga_setelah_diskon = subtotal - diskon
    # Hitung PPN 11% dari harga setelah diskon
    ppn = 0.11 * harga_setelah_diskon
    # Hitung Total Akhir
    total_akhir = harga_setelah_diskon + ppn

    # Output
    print(f"Subtotal: Rp{subtotal:,.0f}")
    print(f"Diskon: Rp{diskon:,.0f}")
    print(f"Harga Setelah Diskon: Rp{harga_setelah_diskon:,.0f}")
    print(f"PPN (11%): Rp{ppn:,.0f}")
    print(f"Total Akhir: Rp{total_akhir:,.0f}")

# Contoh penggunaan:
harga_item = [30000, 45000, 35000]  # Daftar harga item yang dibeli
is_member = True  # Status member
kasir_kedai_kopi(harga_item, is_member)
```

```python
# Fungsi untuk menentukan tahun kabisat
def is_kabisat(tahun):
    # Ekspresi logika kompleks sesuai ketentuan
    return (tahun % 4 == 0 and tahun % 100 != 0) or (tahun % 400 == 0)

# Fungsi input
def input_tahun():
    return int(input("Masukkan tahun: "))

# Fungsi output
def tampilkan_hasil(tahun, hasil):
    if hasil:
        print(f"{tahun} adalah tahun kabisat.")
    else:
        print(f"{tahun} bukan tahun kabisat.")

# Program utama
def main():
    tahun = input_tahun()
    hasil = is_kabisat(tahun)
    tampilkan_hasil(tahun, hasil)

# Menjalankan program
if __name__ == "__main__":
    main()
```
