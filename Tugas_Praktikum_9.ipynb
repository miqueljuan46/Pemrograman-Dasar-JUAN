```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/customers.csv")
```

```python
# Ringkasan statistik
df.describe()
```

```python
# Rata-rata usia
rata_rata_usia = df['age'].mean()
print(rata_rata_usia)

# Modul dari zip_code
modul_zip_code = df['zip_code'].mode()
print(modul_zip_code)
```

```python
df['zip_code'].hist()
plt.title('Frekuensi Zip Code')
plt.xlabel('Zip Code')
plt.ylabel('Frequency')
plt.show()
```

```python
# Kolerasi
corr_matrix = df[['zip_code', 'age']].corr()
print("Korelasi:" , corr_matrix)

# Kovarian
cov_matrix = df[['zip_code', 'age']].cov()
print("Kovarian:" , cov_matrix)
```

```python
KOLOM_GROUP = 'gender'

KOLOM_GROUP = 'gender'

group_mean_data = df.groupby(KOLOM_GROUP).mean(numeric_only=True).reset_index()
print(group_mean_data)


group_agg_data = df.groupby(KOLOM_GROUP).agg({
    'age': ['mean', 'max', 'min'],
    'zip_code': ['mean', 'max', 'min']
})
print(group_agg_data)
```
