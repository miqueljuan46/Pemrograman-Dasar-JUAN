```python
# ATS Praktikum Program Dasar

# 1. Percabangan / Branching
suhu = float(input("Masukkan suhu air pendingin: "))
if suhu <= 90:
    print("STATUS: Normal")
elif suhu < 100:
    print("STATUS: Warning")
else:
    print("ALERT: SHUTDOWN")

# 2. Fungsi / Functions
def hitung_efisiensi(input_energi, output_energi):
    return output_energi / input_energi

hasil_efisiensi = hitung_efisiensi(500, 400)
print("Efisiensi =", hasil_efisiensi)

# 3. Looping for
vibrasi_mesin = [1.2, 0.9, 2.5, 3.1, 1.8]
for i, nilai in enumerate(vibrasi_mesin, start=1):
    if nilai > 2.0:
        print(f"Warning: Unit {i} vibrasi tinggi!")

# 4. If Majemuk
tekanan_uap = float(input("Masukkan tekanan uap: "))
status_manual_override = False
if tekanan_uap > 150 and status_manual_override == False:
    print("Katup Terbuka")

# 5. While Loop
suhu_saat_ini = 60
while suhu_saat_ini < 85:
    print("Suhu saat ini:", suhu_saat_ini)
    suhu_saat_ini += 5

# 6. Variabel & Ekspresi
V = 220
I = 15
daya = V * I
print("Daya listrik =", daya, "Watt")

# 7. Percabangan Beban Listrik
kapasitas = float(input("Masukkan kapasitas maksimum: "))
terpakai = float(input("Masukkan beban terpakai: "))
persen = (terpakai / kapasitas) * 100
if persen <= 60:
    status = "Beban Ringan"
elif persen <= 85:
    status = "Beban Sedang"
else:
    status = "Beban Berat"
print("Status Beban:", status)

# 8. Fungsi Efisiensi Termal
def hitung_efisiensi_termal(energi_masuk, energi_keluar):
    efisiensi = (energi_keluar / energi_masuk) * 100
    print("Efisiensi Termal =", efisiensi, "%")

# 9. Looping Suhu Boiler
suhu_boiler = [92.5, 88.7, 95.1, 102.3, 97.8]
for suhu in suhu_boiler:
    print("Suhu:", suhu)
    if suhu > 100:
        print("Peringatan: Suhu melewati batas aman!")

# 10. While Loop Tangki
volume = 0
while volume <= 100:
    print("Volume sekarang:", volume, "liter")
    if volume == 80:
        print("WARNING: Hampir penuh!")
    if volume == 100:
        print("TANGKI PENUH â€“ Pengisian berhenti.")
        break
    volume += 10

# CHALLENGE PROGRAM

def hitung_efisiensi(input_energi, output_energi):
    return (output_energi / input_energi) * 100

print("=== Monitoring Sistem PLTU ===")
suhu_list = [90, 95, 102, 98]
tekanan = 140
energi_in = 500
energi_out = 420
if tekanan > 150:
    print("Tekanan aman")
else:
    print("Tekanan rendah, naikkan tekanan!")
for s in suhu_list:
    print("Suhu:", s)
    if s > 100:
        print("Warning: Suhu kritis!")
ef = hitung_efisiensi(energi_in, energi_out)
print("Efisiensi Turbin:", ef, "%")
while tekanan < 150:
    tekanan += 2
    print("Tekanan naik:", tekanan)
```
