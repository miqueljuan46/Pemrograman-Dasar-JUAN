```python
# Program menampilkan nama sebanyak 200 kali

# Nama yang akan ditampilkan
nama = "juan Miquel Sinaga"

# Menggunakan perulangan for
print("Menggunakan perulangan for:")
for i in range(200):
  print(f"{i+1}. {nama}")

  # Menggunakan perulangan while
  print("Menggunakan perulangan while:")
  counter = 1
  while counter <= 200:
    print(f"{counter}. {nama}")
    counter += 1
```

```python
# Program menmpilkan angka 1-500 menggunakan while
print("Angka 1 sampai 500 menggunakan while:")
angka = 1
while angka <= 500:
  print(f"{angka:3d}", end=" ")
  if angka % 15 == 0: # Ganti baris setiap 15 angka
    print()
  angka += 1
```

```python
# Menampilkan deret bilangan ganjil 1 sampai 200 menggunakan while

# Inisialisasi variabel
angka = 1

# Perulangan while
while angka <= 200:
  # Cek angka ganjil
  if angka % 2 != 0:
    print(angka, end=" ")
  angka += 1
```
